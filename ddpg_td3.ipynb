### 7. DDPG – Deep Deterministic Policy Gradient
„Deterministische Version von Actor‑Critic für kontinuierliche Aktionen.“

Kernidee
- Off‑policy Actor‑Critic‑Methode für kontinuierliche Aktionsräume.
- Policy ist deterministisch:
$$\pi (s)=a $$
statt einer Wahrscheinlichkeitsverteilung.
- Nutzt Replay Buffer + Target Networks (wie DQN), aber für kontinuierliche Aktionen.

Wesentliche Eigenschaften
- Off‑policy → hohe Sample‑Effizienz.
- Deterministische Policy → effizienter Gradient, aber weniger Exploration.
- Exploration erfolgt über Noise (z. B. Ornstein‑Uhlenbeck).
- Kann instabil sein, da Q‑Funktion leicht überoptimiert wird.

Merksatz
- DDPG ist der „DQN‑Stil“ Actor‑Critic für kontinuierliche Aktionen – schnell, aber empfindlich.


### 8. TD3 – Twin Delayed Deep Deterministic Policy Gradient
„DDPG, aber stabilisiert – weniger Overestimation, weniger Rauschen, bessere Updates.“

Kernidee
- Baut direkt auf DDPG auf, behebt aber dessen Hauptprobleme.
- Drei zentrale Verbesserungen:
- Twin Q‑Networks → reduziert Overestimation (wie Double‑Q).
- Delayed Policy Updates → Actor wird seltener aktualisiert als Critic.
- Target Policy Smoothing → fügt Rauschen zu Target‑Actions hinzu, um Q‑Funktion zu glätten.

Wesentliche Eigenschaften
- Deutlich stabiler als DDPG.
- Robust gegen Overestimation.
- State‑of‑the‑art für deterministische Off‑policy‑Methoden.
- Sehr gute Performance in kontinuierlichen Steuerungsaufgaben.

Merksatz
- TD3 ist DDPG mit drei cleveren Stabilisierungstricks – zuverlässig, robust und performant.
